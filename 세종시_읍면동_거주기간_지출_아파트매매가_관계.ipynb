# 종속변수: 아파트 매매가(2020)

```
apart_mm = pd.read_csv('/content/3.세종시_아파트(매매)_실거래가.csv', encoding='cp949')
apart_mm
```

```
# 동별 아파트 매매가
is_dong = apart_mm['동'].str.endswith('동')
apart_mm_dong = apart_mm[is_dong]
apart_mm_dong[['계약년월', '계약일', '거래금액(만원)', '동']].to_excel('동지역 아파트매매가.xlsx')
apart_mm_dong[['계약년월', '계약일', '거래금액(만원)', '동']].sum()

# 동지역 아파트 매매가 거래금액 총합: 700330557(만원)
```

```
# 읍별 아파트 매매가
is_eup = apart_mm['동']=='조치원읍'
apart_mm_eup = apart_mm[is_eup]
apart_mm_eup[['계약년월', '계약일', '거래금액(만원)', '동']].to_excel('조치원읍지역 아파트매매가.xlsx')
apart_mm_eup[['계약년월', '계약일', '거래금액(만원)', '동']].sum()

# 동지역 아파트 매매가 거래금액 총합: 58365154(만원)
```

```
# 면별 아파트 매매가
is_myeon = apart_mm['동'].str.endswith('면')
apart_mm_myeon = apart_mm[is_myeon]
apart_mm_myeon[['계약년월', '계약일', '거래금액(만원)', '동']].to_excel('면지역 아파트매매가.xlsx')
apart_mm_myeon[['계약년월', '계약일', '거래금액(만원)', '동']].sum()

# 동지역 아파트 매매가 거래금액 총합: 12047808(만원)
```
