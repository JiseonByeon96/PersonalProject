# 데이터 읽어오기

```
# 데이터 읽어오기
import pandas as pd
df = pd.read_csv('/content/아파트매매.csv', encoding='cp949')
df.head()
```

```
df.drop(['계약년월', '계약일', '시군구', '본번', '부번', '단지명', '전용면적(㎡)', '층', '건축년도', '동.1'], axis=1, inplace=True)
```

```
df['계약년월일'] = df['계약년월일'].astype('str')
df['계약년월일'] = pd.to_datetime(df['계약년월일'])
```

```
df['동'].unique()
```

# 보람동

```
is_boram = df['동']=='보람동'
df_boram = df[is_boram].reset_index().drop(['index', '동'], axis=1)
df_boram.columns=['y', 'ds']
df_boram
m = Prophet()
m.fit(df_boram)
future = m.make_future_dataframe(periods=365)
future.tail()
forecast = m.predict(future)
forecast.tail()
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(3)
```

```
fig1 = m.plot(forecast)
fig2 = m.plot_components(forecast)
```
